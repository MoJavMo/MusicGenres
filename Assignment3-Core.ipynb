```python
# importing pandas, and other necessary modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
from sklearn import preprocessing
import warnings
warnings.filterwarnings("ignore")
```

```python
alternative_genre_data = pd.read_csv('~/Downloads/COMP309A3-Data/alternative.csv', index_col=0)
anime_genre_data = pd.read_csv('~/Downloads/COMP309A3-Data/anime.csv', index_col=0)
blues_genre_data = pd.read_csv('~/Downloads/COMP309A3-Data/blues.csv', index_col=0)
classical_genre_data = pd.read_csv('~/Downloads/COMP309A3-Data/classical.csv', index_col=0)
country_genre_data = pd.read_csv('~/Downloads/COMP309A3-Data/country.csv', index_col=0)
electronic_genre_data = pd.read_csv('~/Downloads/COMP309A3-Data/electronic.csv', index_col=0)
hiphop_genre_data = pd.read_csv('~/Downloads/COMP309A3-Data/hip-hop.csv', index_col=0)
jazz_genre_data = pd.read_csv('~/Downloads/COMP309A3-Data/jazz.csv', index_col=0)
rap_genre_data = pd.read_csv('~/Downloads/COMP309A3-Data/rap.csv', index_col=0)
rock_genre_data = pd.read_csv('~/Downloads/COMP309A3-Data/rock.csv', index_col=0)
```

```python
datasets = [alternative_genre_data,
           anime_genre_data,
           blues_genre_data,
           classical_genre_data,
           country_genre_data,
           electronic_genre_data,
           hiphop_genre_data,
           jazz_genre_data,
           rap_genre_data,
           rock_genre_data]
genres = ["alternative",
           "anime",
           "blues",
           "classical",
           "country",
           "electronic",
           "hip-hop",
           "jazz",
           "rap",
           "rock"]
```

```python
mergedData = datasets[0]
i = 0
for data in datasets:
    if i==0:
        i+=1
        continue
    mergedData = mergedData.append(data)
    i+=1
```

```python
newData = mergedData[['loudness', 'energy', 'acousticness', 'instrumentalness']]
poly = preprocessing.PolynomialFeatures(degree=5, interaction_only=True)
newData = poly.fit_transform(newData)
featureCombination = poly.get_feature_names(input_features=['loudness', 'energy', 'acousticness', 'instrumentalness'])
```

```python
featureInteractions = pd.DataFrame(newData, columns=featureCombination)
featureInteractions.pop('1')
featureInteractions.pop('loudness')
featureInteractions.pop('energy')
featureInteractions.pop('acousticness')
featureInteractions.pop('instrumentalness')
```
